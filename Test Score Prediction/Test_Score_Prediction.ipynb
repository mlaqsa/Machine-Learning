```python


import numpy as np
import matplotlib.pyplot as plt


x = np.arange(0, 6, 1)  # Creates an array [0, 1, 2, 3, 4, 5]


y = 10 * x + 50


plt.plot(x, y, label='f(x) = 10x + 50', color='blue')


plt.scatter(x, y, color='red')


plt.xlabel('Hours Studied')
plt.ylabel('Test Score')
plt.title('Linear Function: Test Score Prediction')
plt.grid(True)
plt.legend()


plt.show()
```
